```python
import faiss
import numpy as np

d = 64
index = faiss.IndexFlatL2(d)
# gpu_res = faiss.StandardGpuResources()  # 初始化GPU资源
# gpu_index = faiss.index_cpu_to_gpu(gpu_res, 0, index)  # 将索引转移到GPU

xb = np.random.random((10000, d)).astype('float32')  # 生成一些随机数据
index.add(xb)  # 将数据添加到索引

xq = np.random.random((1, d)).astype('float32')  # 创建一个查询向量
k = 4  # 我们想要查找最近的4个向量
D, I = index.search(xq, k)  # 进行搜索
print(I)  # 打印出最近邻的索引
print(D)
```
